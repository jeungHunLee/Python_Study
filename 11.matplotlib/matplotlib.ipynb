```python
"""선 그래프 그리기"""
import matplotlib.pyplot as plt

data = [50, 125, 300, 150, 400]

plt.plot(data)
plt.show()
```

```python
"""곡선 그래프 그리기"""
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
"""막대 그래프 그리기"""
import matplotlib.pyplot as plt

data = [50, 125, 300, 150, 400]

plt.bar(range(len(data)), data)
plt.show()
```
