```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

t1 = np.arange(0., 5., 0.1)
t2 = np.arange(0., 5., 0.02)

plt.figure(1)
plt.subplot(211)    # 2행 1열 1번
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(2, 1, 2)    # 2행 1열 2번
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# GridSpec: subplot이 들어갈 위치의 그리드
grid = plt.GridSpec(2, 3, wspace=0.5, hspace=0.3)    # nrows, ncols, 좌우간격, 상하간격
ax1 = plt.subplot(grid[0, 0])
ax2 = plt.subplot(grid[0, 1:])
ax3 = plt.subplot(grid[1, :2])
ax4 = plt.subplot(grid[1, 2])

ax1.plot(np.arange(10))
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
ax2.plot(np.sin(x))
ax3.plot(np.cos(x))
ax4.bar(np.arange(10), np.random.randint(0, 100, 10))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0., 5.)    # default 50개
x2 = np.linspace(0., 2.)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

fig, (ax1, ax2) = plt.subplots(2, 1)    # 2행 1열에 대한 figure 객체와 axes 반환
fig.suptitle('Title')    # 메인 타이틀

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Damped oscillation')

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('time(s)')
ax2.set_ylabel('Undamped')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x ** 2 + y ** 2

x, y = np.meshgrid(np.linspace(-10, 10, 100), np.linspace(-10, 10, 100))    # mesh grid
z = f(x, y)

plt.contourf(x, y, z, level=15)    # x, y 좌표평면, z: 높이 값, level: 몇 단계로 나타낼 것인지
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)    # mesh grid
Z = f(X, Y)

plt.contourf(X, Y, Z, level=20, cmap='RdGy')    # level: 등고선 영역 수
plt.colorbar()
plt.show()
```
